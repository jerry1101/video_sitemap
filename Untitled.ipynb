```python
import requests
import json

class BearerAuth(requests.auth.AuthBase):
    token = None
    def __init__(self, token):
        self.token = token
    def __call__(self, r):
        r.headers["authorization"] = "Bearer " + self.token
        return r
```

```python
vimeo_response = requests.get('https://api.vimeo.com/videos/138084222', auth=BearerAuth('46bdef9366425b4caf065943d3d46e01'))
```

```python
import pprint
import json
def pp_json(json_thing, sort=True, indents=4):
    if type(json_thing) is str:
        print(json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents))
    else:
        print(json.dumps(json_thing, sort_keys=sort, indent=indents))
    return None
pp_json(vimeo_response.text)
```

```python
data = vimeo_response.text
video=json.loads(data)
print(video['description'])
print(video['name'])
print(video['link'])
print(video['pictures']['sizes'][1]['link'])
```
