```python
import requests
response = requests.get('http://localhost:8080/hybrisquery',data= 'select distinct {code},{VIDEOURL} from {PRODUCT} where {VIDEOURL} is not null')
print(response.text)
```

```python
import pandas as pd
# print(response.text)
data = pd.read_json(response.text)
data.rename(columns={'code':'Product_Code',
                          'p_videourl':'Video_ID',
                          }, 
                 inplace=True) 
data
```

```python
data.to_html('test.html')
```

```python
data.to_csv('test.csv', sep='\t', encoding='utf-8')
```
